```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Taking required data
re=3000 #ft
rwf=0.5 #ft
pwf=1500 #psi
h=60 #ft
k=150 #md
q=200#stb/d
mu=15 #cp
B=1 #rsb/stb
#steady state reservoir (supported by strong aquifier)
```

```python
#array of rdius
r=np.linspace(rwf,re,500)  # creating 500 points in between in rw and re
```

```python
#Equation of radial Flow of Incompressible Fluids
#Q=2*PI*k*h(ln(re/rw))*(Pe-Pw)/(B*mu*l)
```

```python
#Calculation of pressure at different point at reservoir
pressure=[]
for i in range(len(r)):
  p= pwf+(141.2*q*mu*B*(np.log(r[i]/rwf)/(k*h)))
  pressure.append(p)
```

```python
#Plotting the curves
plt.style.use("classic")
plt.figure(figsize=[8,6])
plt.plot(r,pressure,lw=3)
plt.xlabel('Radius (ft)',size=15)
plt.ylabel('Pressure (psi)',size=15)
plt.title('Resevoir pressure profile',size=15)

plt.grid(True)
```
