```python
# Import some libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating 100 data of water saturation in between 0 and 1
sw=np.linspace(0,1,100)

#k=a*sw**b
# let a=1 and b=3
```

```python
#Calculating relative permeability
krw=1*sw**3 ; kro=(1-sw)**3
```

```python
#Plotting
plt.style.use("default")
plt.plot(sw,krw,label="Relative Permeability of Water",lw=5)
plt.plot(sw,kro,label="Relative Permeability of Oil",lw=5)
plt.xlabel("Water Saturation")
plt.ylabel("Relative Permeability")
plt.title("Relative Permeability of Water and Oil")
plt.legend()
plt.grid()
plt.show()
```
